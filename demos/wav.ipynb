```python
# import os module
import os

# import AUTO modules
from auto import AUTOCommands as ac
from auto import AUTOclui as acl
from auto import interactiveBindings as ib
from auto import runAUTO as ra

# import plotting tools
import matplotlib as mpl
from matplotlib import pyplot as plt

# virtual display to use AUTO plot function
from pyvirtualdisplay import Display
from IPython.display import display as display
from PIL import Image
disp = Display(visible=0, size=(1366,768))
disp.start()
```

```python
# use working directory for demo to avoid clutter
folder = '/auto/workspace/demos/demo'
os.chdir(folder)
```

```python
# copy demo into current directory
ac.copydemo('wav')

# start AUTO with runner object
runner = ra.runAUTO()

pgm = "wav"
print("\nDemo %s is started"%pgm)
print("\n%s : first run : stationary solutions of the ODE"%pgm)
r1 = ac.run(e=pgm,c=pgm,runner=runner)
ac.save(r1,"ode")

print("\n%s : second run : stationary solutions of the reduced PDE"%pgm)
r2 = ac.run(e=pgm,c=pgm,IPS=11,runner=runner)

print("\n%s : third run : wave train solutions of fixed wave speed"%pgm)
r3 = ac.run(r2("HB1"),IPS=12,ICP=[3,11],ILP=0,ISP=0,RL1=700,DS=0.1,DSMAX=1.0,
         UZR={3:[610.0,638.0], 11:500.0}, STOP=['UZ4'],runner=runner)
ac.save(r2+r3,pgm)

print("\n%s : fourth run : fixed wave length waves on a ring"%pgm)
uz3 = ac.load(r3("UZ3"),RL1=1000,DS=0.5,DSMAX=2.0,UZR={},runner=runner)
r4 = ac.run(uz3,ICP=[3,10],NPR=50,runner=runner)
ac.save(r4,'rng')

print("\n%s : fifth run : time evolution computation"%pgm)
r5 = ac.run(uz3,IPS=14,ICP=[14],NMX=230,NPR=5,runner=runner)
ac.save(r5,'tim')

#print("\n%s : sixth run : stationary wave continuation"%pgm)
#r6 = ac.run(r2+r3,c=pgm+'6',runner=runner)
#ac.save(r2+r3+r6,pgm)
```

```python
# plot bifurcation diagram using plot routine from AUTO
p = ac.plot('wav')
p.config(stability=True, grid=False)
p.savefig('b.png')
display(Image.open('b.png'))

# wave train solutions of fixed wave speed
p = ac.plot('wav')
p.config(type='solution')
p.config(solution_x="t", solution_y="U(1)")
p.config(label=list(range(6,10)))
p.config(grid=False)
p.savefig('b.png')
display(Image.open('b.png'))

# plottime evolution of solution
p = ac.plot('tim')
p.config(type='solution')
p.config(solution_x="U(1)", solution_y="U(2)")
p.config(grid=False)
p.savefig('b.png')
display(Image.open('b.png'))
```
