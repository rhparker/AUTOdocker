```python
# import os module
import os

# import AUTO modules
from auto import AUTOCommands as ac
from auto import AUTOclui as acl
from auto import interactiveBindings as ib
from auto import runAUTO as ra

# import plotting tools
import matplotlib as mpl
from matplotlib import pyplot as plt

# virtual display to use AUTO plot function
from pyvirtualdisplay import Display
from IPython.display import display as display
from PIL import Image
disp = Display(visible=0, size=(1366,768))
disp.start()
```

```python
# use working directory for demo to avoid clutter
folder = '/auto/workspace/demos/demo'
os.chdir(folder)
```

```python
# copy demo into current directory
ac.copydemo('int')

# start AUTO with runner object
runner = ra.runAUTO()

pgm = "int"
print("\nDemo %s is started"%pgm)
print("\n%s : first run : detection of a fold"%pgm)
r1 = ac.run(e=pgm,c=pgm,runner=runner)
ac.save(r1,pgm)

print("\n%s : second run : generate starting data"%pgm)
r2 = ac.run(r1("LP1"),ICP=[1,2],ISW=2,runner=runner)

print("\n%s : third run : compute a curve of folds"%pgm)
r3 = ac.run(r2,runner=runner)
ac.save(r3,'lp')
```

```python
# plot bifurcation diagrams using plot routine from AUTO

p = ac.plot('int')
p.savefig('b.png')
display(Image.open('b.png'))

# par(2) vs par(1) for curve of folds
p = ac.plot('lp')
p.config(bifurcation_x="PAR(1)", bifurcation_y="PAR(2)")
p.savefig('b.png')
display(Image.open('b.png'))

```
