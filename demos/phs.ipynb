```python
# import os module
import os

# import AUTO modules
from auto import AUTOCommands as ac
from auto import AUTOclui as acl
from auto import interactiveBindings as ib
from auto import runAUTO as ra

# import plotting tools
import matplotlib as mpl
from matplotlib import pyplot as plt

# virtual display to use AUTO plot function
from pyvirtualdisplay import Display
from IPython.display import display as display
from PIL import Image
disp = Display(visible=0, size=(1366,768))
disp.start()
```

```python
# use working directory for demo to avoid clutter
folder = '/auto/workspace/demos/demo'
os.chdir(folder)
```

```python
# copy demo into current directory
ac.copydemo('phs')

# start AUTO with runner object
runner = ra.runAUTO()

print("\n***Compute stationary solution families***")
r1 = ac.run(e='phs',c='phs.1',runner=runner)

print("\n***Compute a periodic solution family***")
r2 = ac.run(r1('HB1'),c='phs.2',runner=runner)
ac.save(r1+r2,'phs')
```

```python
# plot periodic orbits

# U(1) vs t
p = ac.plot('phs')
p.config(type='solution')
p.config(grid=False)
p.config(label=[10,15,20,25])
p.savefig('b.png')
display(Image.open('b.png'))

# U(2) vs U(1)
p.config(solution_y="U(2)")
p.config(solution_x="U(1)")
p.savefig('b.png')
display(Image.open('b.png'))
```
